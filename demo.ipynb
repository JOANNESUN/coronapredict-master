```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### 資料來源：https://www.kaggle.com/sudalairajkumar/novel-corona-virus-2019-dataset

```python
data = pd.read_csv('20200318/covid_19_data.csv')
```

```python
data.describe()
```

```python
data.head(50)
```

```python
data['ObservationDate'] = pd.to_datetime(data['ObservationDate'])
data
```

```python
world = data.groupby('ObservationDate').sum()
world
```

```python
world = world['Confirmed']
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 16, 9
plt.plot(world)
```

```python
from pyramid.arima import auto_arima
model = auto_arima(world)
model.aic()
```

```python
train = world.loc['2020-01-22':'2020-02-29']
test = world.loc['2020-03-01':]
print(train.shape)
print(test.shape)
```

```python
model.fit(train)
```

```python
predict = model.predict(n_periods=18)
predict
```

```python
predict = pd.DataFrame(predict, index = test.index, columns=['Predict'])
df = pd.concat([test, predict], axis = 1)
df
```
