```python
from sklearn.model_selection import train_test_split

X = AU_series2.values # turn data into array 

# split data into training and test data.
train_X, test_X = train_test_split(X,train_size=0.7, test_size=0.3, random_state=123)
predictions = []
```

```python
train_X.size
```

```python
test_X.size
```

```python
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
model_arima = ARIMA(train_X,order =(1, 1, 1))
model_arima_fit = model_arima.fit()
print(model_arima_fit.aic) # the lower the better 
```

```python
predictions = model_arima_fit.forecast(steps=17)[0]
predictions
```

```python
plt.plot(test_X)
plt.plot(predictions, color = 'red')
```

#Simple function that finds the least AIC number using p-d-q parameter
import warnings
warnings.filterwarnings('ignore')
import itertools
p=d=q=range(0,6)
pdq = list(itertools.product(p,d,q))

for parameter in pdq:
    try:
        model_arima1 = ARIMA(train_X,order =parameter)
        model_arima1_fit = model_arima1.fit()
        print(parameter, model_arima1_fit.aic)
    except:
        continue
